```python
import numpy as np
from arc.alkali_atom_data import Rubidium85, Rubidium87
from arc import LevelPlot
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='retina'
```

```python
rb85_atom = Rubidium85()
# rb_atom.getEnergy(n=5, l=0, j=1/2)  # eV

levels85 = LevelPlot(rb85_atom)
levels85.makeLevels(nFrom=5, nTo=5, lFrom=0, lTo=1, sList=[1/2])
levels85.drawLevels('GHz')
levels85.showPlot()

rb87_atom = Rubidium87()
# rb_atom.getEnergy(n=5, l=0, j=1/2)  # eV

levels87 = LevelPlot(rb87_atom)
levels87.makeLevels(nFrom=5, nTo=5, lFrom=0, lTo=1, sList=[1/2])
levels87.drawLevels('GHz')
levels87.showPlot()
```
